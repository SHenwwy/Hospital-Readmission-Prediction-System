```python
cat = df.select_dtypes(include=['object']).columns
num = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
```

```python
cat_features = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

num_features = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
preprocessing = ColumnTransformer(transformers=[
    ('cat', cat_features, cat),
    ('num', num_features, num)
])
```

```python
le = LabelEncoder()
target = le.fit_transform(df['readmitted'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=42)
```

```python
end = Pipeline(steps=[
    ('preprocessing', preprocessing),
    ('model', LogisticRegression())
])
```

```python
scores = cross_val_score(end, X_train, y_train, cv=5)

print(f"Average Cross-Validation Accuracy: {scores.mean()}")
```

```python
end.fit(X_train, y_train)
```

```python
test_score = end.score(X_test, y_test)
print(f"Test Set Accuracy: {test_score}")
```

```python
predicts = end.predict(X_test)
```

```python
accuracy_score(y_test, predicts)
```

```python
report = classification_report(y_test, predicts)
print(report)
```
